```python
import numpy as np

def find_outliers(data):
    # 计算平均值和标准差
    mean = np.mean(data)
    std = np.std(data)
    print("Mean:", mean)
    print("Standard deviation:", std)
    # 定义一个阈值，这里我们设定为平均值加上两倍标准差
    threshold = mean + 2 * std
    
    # 找到大于阈值的数
    outliers = [x for x in data if x > threshold]
    
    return outliers

# 示例数据
data = [100,34,0.7]

# 找到离群值
outliers = find_outliers(data)

print(outliers)
```

```python
#计算f1值

list_all = [1000,1000,1000]
list_0 = [779,200,21]
list_1 = [160,832,8]
list_2 = [140,7,853]

#上面的list_0，list_1，list_2分别是三个类别在0，1，2类上的预测结果，list_all是0，1，2三个类别的样本数，计算f1值
def f1_score_3(list_all,list_0,list_1,list_2):
    precision_0 = list_0[0]/(list_0[0]+list_1[0]+list_2[0])
    recall_0 = list_0[0]/list_all[0]
    f1_0 = 2*precision_0*recall_0/(precision_0+recall_0)
    precision_1 = list_1[1]/(list_0[1]+list_1[1]+list_2[1])
    recall_1 = list_1[1]/list_all[1]
    f1_1 = 2*precision_1*recall_1/(precision_1+recall_1)
    precision_2 = list_2[2]/(list_0[2]+list_1[2]+list_2[2])
    recall_2 = list_2[2]/list_all[2]
    f1_2 = 2*precision_2*recall_2/(precision_2+recall_2)
    f1 = (f1_0+f1_1+f1_2)/3
    return f1


f1 = f1_score_3(list_all,list_0,list_1,list_2)
print(f1)


list_all = [1000,1000]
list_0 = [779,221]
list_1 = [168,832]

#上面的list_0，list_1分别是两个类别在0，1类上的预测结果，list_all是0，1两个类别的样本数，计算f1值
def f1_score_2(list_all,list_0,list_1):
    precision_0 = list_0[0]/(list_0[0]+list_1[0])
    recall_0 = list_0[0]/list_all[0]
    f1_0 = 2*precision_0*recall_0/(precision_0+recall_0)
    precision_1 = list_1[1]/(list_0[1]+list_1[1])
    recall_1 = list_1[1]/list_all[1]
    f1_1 = 2*precision_1*recall_1/(precision_1+recall_1)
    f1 = (f1_0+f1_1)/2
    return f1
f1 = f1_score_2(list_all,list_0,list_1)
print(f1)
```

```python
# 导入TfidfVectorizer类
from sklearn.feature_extraction.text import TfidfVectorizer

# 定义一组字符串
# documents = [
#     'the car is driven on the road',
#     'the truck is driven on the highway'
# ]

s1 = 'the Car is driven on the road'
s2 = 'the truck is dRiven on the highway'

# #转换成小写
# s1 = s1.lower()


from gensim.parsing.preprocessing import STOPWORDS 
import numpy 
stop_words = STOPWORDS 
# 定义一个文本字符串 
def filter(tokens):
    filtered_tokens = [token for token in tokens if token not in stop_words and len(token) > 1] 
    #拼成字符串
    filtered_tokens = ' '.join(filtered_tokens)
    #小写
    filtered_tokens = filtered_tokens.lower()
    return filtered_tokens

s1 = filter(s1.split())
s2 = filter(s2.split())


documents = [s1, s2]

def tfidf(documents,index):
    # 初始化一个TfidfVectorizer对象
    vectorizer = TfidfVectorizer()
    # 调用fit_transform方法计算TF-IDF值
    tfidf_matrix = vectorizer.fit_transform(documents)
    #输出第一句话的单词和对应的TF-IDF值
    # print(vectorizer.get_feature_names())
    # print(tfidf_matrix.toarray()[0])
    #构建字典来存储每个单词的TF-IDF值
    word2tfidf = dict(zip(vectorizer.get_feature_names(), tfidf_matrix.toarray()[index]))
    # print(word2tfidf)
    #排序, 按照TF-IDF值从大到小排序
    sorted_word2tfidf = sorted(word2tfidf.items(), key=lambda x: x[1], reverse=True)
    # print(sorted_word2tfidf)
    return sorted_word2tfidf

sorted_word2tfidf = tfidf(documents,0)
print(sorted_word2tfidf)
sorted_word2tfidf = tfidf(documents,1)
print(sorted_word2tfidf)
# # 查看结果矩阵
# matix = tfidf_matrix.toarray()
# for i in matix:
#     print(i)
#     print('-----------------')
# #取出第一行

# #排序，按照TF-IDF值从大到小排序
# # 查看特征名称（即词语）
# print(vectorizer.get_feature_names())

```

```python
import threading

def sum_list(lst):
    return sum(lst)

def split_list(lst, n):
    k, m = divmod(len(lst), n)
    return (lst[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(n))

lst = list(range(100))
threads = []
for i, sub_lst in enumerate(split_list(lst, 10)):
    t = threading.Thread(target=lambda x: print(f"Thread {x}: {sum_list(sub_lst)}"), args=(i,))
    threads.append(t)
    t.start()

for t in threads:
    t.join()
```

```python
# 导入相关库
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# 定义文本集合和单词集合
D = ["Human machine interface for lab abc computer applications",
     "A survey of user opinion of computer system response time",
     "The EPS user interface management system",
     "System and human system engineering testing of EPS",
     "Relation of user perceived response time to error measurement",
     "The generation of random binary unordered trees",
     "The intersection graph of paths in trees",
     "Graph minors IV Widths of trees and well quasi ordering",
     "Graph minors A survey"]

W = ["human", "interface", "computer", "user", "system", "response", 
"time", "EPS", "survey", "trees", "graph", "minors"]

# 构建TF-IDF矩阵
vectorizer = TfidfVectorizer(vocabulary=W)
X = vectorizer.fit_transform(D).toarray()

# 对TF-IDF矩阵进行奇异值分解
U, S, Vt = np.linalg.svd(X)

# 选择前两个奇异值，截断奇异值分解
k = 2
Uk = U[:, :k]
Sk = np.diag(S[:k])
Vtk = Vt[:k, :]

# 得到文档-主题矩阵和主题-术语矩阵
DT_matrix = Uk @ Sk # 文档-主题矩阵，每一行代表一个文档在两个主题上的相关程度
TT_matrix = Sk @ Vtk # 主题-术语矩阵，每一行代表一个主题在所有术语上的相关程度

print("文档-主题矩阵：")
print(DT_matrix)
print("主题-术语矩阵：")
print(TT_matrix)
```
